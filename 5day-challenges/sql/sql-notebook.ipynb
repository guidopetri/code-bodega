# Day 2

```python
# using group by, having, and aggregate functions like count/avg

db_connection_string = 'postgresql://{}:{}@{}:{}/{}'

with psycopg2.connect(db_connection_string.format(user,
                                                  password,
                                                  ip,
                                                  port,
                                                  dbname)) as con:
    cursor = con.cursor()

    sql = """SELECT player, count(opponent) from chess_games
                     group by player;
                  """
    cursor.execute(sql)
    colnames = [desc.name for desc in cursor.description]
    results = cursor.fetchall()
df = pd.DataFrame.from_records(results, columns=colnames)
df.head()
```

```python
db_connection_string = 'postgresql://{}:{}@{}:{}/{}'

with psycopg2.connect(db_connection_string.format(user,
                                                  password,
                                                  ip,
                                                  port,
                                                  dbname)) as con:
    cursor = con.cursor()

    sql = """SELECT player, opponent, count(game_link) from chess_games
                     group by player, opponent
                     having count(game_link) > 1;
                  """
    cursor.execute(sql)
    colnames = [desc.name for desc in cursor.description]
    results = cursor.fetchall()
df = pd.DataFrame.from_records(results, columns=colnames)
df.head()
```

# Day 3

```python
db_connection_string = 'postgresql://{}:{}@{}:{}/{}'

with psycopg2.connect(db_connection_string.format(user,
                                                  password,
                                                  ip,
                                                  port,
                                                  dbname)) as con:
    cursor = con.cursor()

    sql = """SELECT player, avg(coalesce(player_rating_diff, 0)) from chess_games
                     group by player
                     order by avg desc;
                  """
    cursor.execute(sql)
    colnames = [desc.name for desc in cursor.description]
    results = cursor.fetchall()
df = pd.DataFrame.from_records(results, columns=colnames)
df.head()
```

```python
db_connection_string = 'postgresql://{}:{}@{}:{}/{}'

with psycopg2.connect(db_connection_string.format(user,
                                                  password,
                                                  ip,
                                                  port,
                                                  dbname)) as con:
    cursor = con.cursor()

    sql = """SELECT player, avg(player_rating_diff),
                    to_char(datetime_played, 'Dy') as day_of_week
                    from chess_games
                    where player_rating_diff != 'nan' and player = 'Grahtbo'
                    group by player, day_of_week
                    order by avg desc;
                  """
    cursor.execute(sql)
    colnames = [desc.name for desc in cursor.description]
    results = cursor.fetchall()
df = pd.DataFrame.from_records(results, columns=colnames)
df.head(10)
```

# Day 4

```python
db_connection_string = 'postgresql://{}:{}@{}:{}/{}'

with psycopg2.connect(db_connection_string.format(user,
                                                  password,
                                                  ip,
                                                  port,
                                                  dbname)) as con:
    cursor = con.cursor()

    sql = """WITH tbo_games as (
                SELECT * from chess_games
                    where player_rating_diff != 'nan' and player = 'Grahtbo'
                    )
    
                SELECT player, avg(player_rating_diff),
                    to_char(datetime_played, 'Dy') as day_of_week
                    from tbo_games
                    group by player, day_of_week
                    order by avg desc;
                  """
    cursor.execute(sql)
    colnames = [desc.name for desc in cursor.description]
    results = cursor.fetchall()
df = pd.DataFrame.from_records(results, columns=colnames)
df.head(10)
```

# Day 5

```python
db_connection_string = 'postgresql://{}:{}@{}:{}/{}'

with psycopg2.connect(db_connection_string.format(user,
                                                  password,
                                                  ip,
                                                  port,
                                                  dbname)) as con:
    cursor = con.cursor()

    sql = """WITH tbo_games as (
                SELECT * from chess_games
                left join eco_codes on chess_games.opening_played = eco_codes.eco_code
                    where player_rating_diff != 'nan' and player = 'Grahtbo'
                    )
    
                SELECT player, avg(player_rating_diff),
                    eco_code, opening_name
                    from tbo_games
                    group by player, eco_code, opening_name
                    order by avg desc;
                  """
    cursor.execute(sql)
    colnames = [desc.name for desc in cursor.description]
    results = cursor.fetchall()
df = pd.DataFrame.from_records(results, columns=colnames)
df.head(10)
```
