```python
import bs4
import pandas as pd
import requests
```

```python
url = 'https://log.concept2.com/rankings/2023/rower/100?rower=rower&weight=H&gender=F'

r = requests.get(url)
r.status_code
```

```python
soup = bs4.BeautifulSoup(r.text, 'html5lib')

tables = soup.find_all(class_='table')

len(tables)

results = []

for row in tables[1].find('tbody').find_all('tr'):
    result = [td.contents[0] if td.contents else None for td in row.find_all('td')]
    # unwrap name
    result.append(result[1].get('href')[40:])
    result[1] = result[1].contents[0]
    results.append(result)

df = pd.DataFrame.from_records(results,
                               columns=['standing', 'name', 'age',
                                        'location', 'country', 'org', 'time', 'verified',
                                        'profile_link',
                                        ])
```
